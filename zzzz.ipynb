```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('zomato.csv',encoding='latin-1')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country
```

```python
fd=pd.merge(df,df_country,on='Country Code',how='left') 
```

```python
fd
```

```python
fd.Country.value_counts()
```

```python
country_val=fd.Country.value_counts().values 
```

```python
country_names=fd.Country.value_counts().index 
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
fd.columns
```

```python
ratings=fd.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['black','blue','white','yellow'])
plt.figure(figsize=(10,10))
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['black','blue','white','yellow'])
```
